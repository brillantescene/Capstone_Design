```python
import os
import natsort
```

```python
def write_lst(file_path, lst_name):
    file_lists = os.listdir(file_path)
    file_lists = natsort.natsorted(file_lists)
    f = open(f'{lst_name}.lst', 'w', encoding='utf-8')
    for idx, file_name in enumerate(file_lists):
        if "drop" in file_name:
            f.write(f"{idx}\t0\t{file_name}\n")
        elif "motor" in file_name:
            f.write(f"{idx}\t1\t{file_name}\n")
        else:
            f.write(f"{idx}\t2\t{file_name}\n")
    f.close()
```

```python
write_lst('train/', 'train')
```

```python
write_lst('valid/', 'val')
```
